#### uploading a video file and load the pre-trained YOLOv5 model for object detection


```python
from google.colab import drive
drive.mount('/content/drive')

# Specify the path to the video file in Google Drive
video_path = '/content/drive/MyDrive/video_mobil.mp4'

```

```python
import torch

# Load YOLOv5 model (small version for efficiency)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
import cv2

# Initialize video capture with the video file
video_capture = cv2.VideoCapture(video_path)

# Check if video opened successfully
if not video_capture.isOpened():
    print("Error: Could not open video file.")

# Define parameters for optimization
frame_skip = 5             # Process every 5th frame
resize_dim = (640, 360)    # Resize to this resolution
frame_count = 0

# Process each frame in the video
while video_capture.isOpened():
    ret, frame = video_capture.read()
    if not ret:
        break  # End of video

    # Skip frames to save processing time
    if frame_count % frame_skip == 0:
        # Resize frame to reduce load
        frame = cv2.resize(frame, resize_dim)

        # Detect objects in the current frame
        results = model(frame)

        # Extract bounding box coordinates, class labels, and confidence scores
        detections = []
        for *box, conf, cls in results.xyxy[0]:  # Format: [x1, y1, x2, y2, confidence, class]
            detection_data = {
                'bounding_box': [int(coord) for coord in box],  # Bounding box as [x1, y1, x2, y2]
                'confidence': float(conf),
                'class': model.names[int(cls)],  # Class name
                'class_id': int(cls)  # Class ID
            }
            detections.append(detection_data)

        # Print detections for every processed frame (optional)
        print(f"Detections in Frame {int(video_capture.get(cv2.CAP_PROP_POS_FRAMES))}: {detections}")

    frame_count += 1

# Release the video capture object
video_capture.release()
```

#### Optimazing YOLO Performance

```python
import cv2
import time

# Path for input video file
video_path = '/content/drive/MyDrive/video_mobil.mp4'   # Update this with your actual input video path
# Path for output video file
output_path = '/content/drive/My Drive/output_video_YOLO_tugas.mp4'

# Initialize video capture
video_capture = cv2.VideoCapture(video_path)

if not video_capture.isOpened():
    print("Error: Could not open video file.")

# Get video properties
fps = video_capture.get(cv2.CAP_PROP_FPS)  # Frames per second
width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))  # Frame width
height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))  # Frame height

# Initialize video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for mp4
video_writer = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

frame_skip = 1
resize_dim = (640, 360)  # Resize frames for faster processing
frame_count = 0

while video_capture.isOpened():
    ret, frame = video_capture.read()
    if not ret:
        break

    if frame_count % frame_skip == 0:
        # Resize the frame for processing
        frame_resized = cv2.resize(frame, resize_dim)

        # Measure processing time
        start_time = time.time()

        # Detect objects
        results = model(frame_resized)  # Detect objects

        # Process detections
        detections = results.xyxy[0].tolist()  # Extract detections

        # Calculate scaling factors
        height_scale = height / resize_dim[1]
        width_scale = width / resize_dim[0]

        # Draw bounding boxes on the original frame
        for *box, conf, cls in detections:
            x1, y1, x2, y2 = map(int, box)  # Convert to int

            # Adjust bounding box coordinates to original frame size
            x1 = int(x1 * width_scale)
            y1 = int(y1 * height_scale)
            x2 = int(x2 * width_scale)
            y2 = int(y2 * height_scale)

            label = f"{model.names[int(cls)]} {conf:.2f}"  # Class label with confidence
            # Draw rectangle on the original frame
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)  # Draw label

        # Write the processed frame to the output video
        video_writer.write(frame)  # Write the original frame with detections

        end_time = time.time()
        print(f"Frame {frame_count} processed in {end_time - start_time:.2f} seconds")

    frame_count += 1

# Release resources
video_capture.release()
video_writer.release()  # Close the video writer
print(f"Output video saved to {output_path}")
```
